```python
y = np.random.standard_normal((20, 2)).cumsum(axis=0)
```

```python
plt.figure(figsize=(10,6))
plt.plot(y, lw=1.5)
plt.plot(y, 'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot');
```

```python
plt.figure(figsize=(10,6))
plt.plot(y[:, 0], lw=1.5, label = '1st')
plt.plot(y[:, 1], lw=1.5, label = '2nd')
plt.plot(y, 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot');
```
